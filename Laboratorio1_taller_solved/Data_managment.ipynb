##Regresion lineal(50_startups dataset) (50%)

Descripción del dataset: \
El dataset contiene datos de 50 startups respecto a algunos gastos y ganancias
El dataset contiene lo siguiente:
* R&D spend: Gasto de la empresa en desarrollo e investigación 
* Administration: Gasto de la empresa en administracin
* Marketing spend: Gasto de la empresa en marketing
* State: Ubicación de la empresa
* Profit: Ganancias de la empresa

```python
#Cargar el datset
ds=pd.read_csv('50_Startups.csv')
ds
```

```python
#Hacer una inspección de este, ver cuantos valores nulos hay, la descripción y la correlación entre sus variables

ds.isnull().sum()
sns.heatmap(data=ds.corr(),annot=True)
#ds.describe()
#ds.corr()
```

```python
#Usar la media para rellenar aquellos valores que se encuentren nulos
med=ds.median()
ds1=ds.fillna(med)
ds1
```

```python
#Aplicar OneHotEncoding a la varible State

from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder

ct=make_column_transformer([OneHotEncoder(sparse=False), ['State']], remainder='passthrough')
ds1_ct=ct.fit_transform(ds1)

ds1_ct= pd.DataFrame(data=ds1_ct, columns=ct.get_feature_names_out())
ds1_ct
```

```python
#Seprar el dataset, la variable dependiente(target) sera 'Profit', las demas seran independientes
from sklearn.model_selection import train_test_split

X=ds1_ct.drop('remainder__Profit', axis=1)
y=ds1_ct['remainder__Profit']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=33)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=1)

print(f'train: {X_train.shape}')
print(f'test: {X_test.shape}')
print(f'val: {X_val.shape}')
```

```python
name_columns=X_train.columns
```

```python
#dividir el dataset en entrenamiento y prueba, donde entrenamiento tendrá el 75% de los datos y el random_state sea 33
```

```python
#Estandarizar el dataset ustando StandarScaler
from sklearn.preprocessing import StandardScaler, Normalizer
sc = StandardScaler()
X_train = sc.fit_transform(X_train.values)

X_val= sc.transform(X_val.values)

X_test = sc.transform(X_test.values)
```

```python
X_train = pd.DataFrame(data=X_train, columns=name_columns)
X_test = pd.DataFrame(data=X_test, columns=name_columns)
X_val = pd.DataFrame(data=X_val, columns=name_columns)
```

```python
#Entrenar el modelo de regresion lineal

from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
#realizar predicción en el grupo de test y comprar resultados entre los valores reales y los predichos
y_pred = linear_model.predict(X_test)
```

```python
results = pd.DataFrame(data={'y_pred':y_pred, 'y_real':y_test})

results
```

```python
#observar los valores de los coeficientes y analizar cual de las variables independientes tiene mas peso

linear_model.coef_
linear_model.feature_names_in_


```

```python
#sacar las metricas de r2_score, error cuadratico medio y error absoluto medio

'''
Valores aproximados
Error cuadratico medio: 159713330.6580093
Error absoluto medio: 10034.99719775062
r2 score: 0.8698765436584724
'''

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)

print(f'Error cuadratico medio: {mse}')

mae = mean_absolute_error(y_test, y_pred)
print(f'Error absoluto medio: {mae}')

rtwo_score = r2_score(y_test, y_pred)
print(f'r2 score: {rtwo_score}')
```

```python
#graficar resultados y conclusiones del modelo
plt.figure(figsize = (15,8))
ax = sns.barplot(linear_model.feature_names_in_,linear_model.coef_ )
ax.set_xticklabels(ax.get_xticklabels(), rotation = 30)
plt.show()
```
